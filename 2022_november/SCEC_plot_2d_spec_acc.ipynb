```python
%matplotlib notebook
import sys
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

##
def read_dataframe(picklename=None):
    try:
        df1 = pd.read_pickle(picklename)
        df = df1.copy()
        print ('df read:', picklename)
    except:
        import pickle5 as pickle
        with open(picklename, "rb") as fh:
            df1 = pickle.load(fh)
            df = df1.copy()
    return df
##
```

```python
df_model1 = read_dataframe(picklename='./store/DF_Model1')
df_model2 = read_dataframe(picklename='./store/DF_Model2')
df_model3 = read_dataframe(picklename='./store/DF_Model3')
```

# 2D plot

## Model 1

```python
df = df_model1
A = np.array([-28.7, 0.0])
B = np.array([20.9, 0.0])
x_hypo, z_hypo = 1.45, -7.0

###
plt.close('all')
fig = plt.figure(figsize=(6,3))
#
ax = fig.add_subplot(111)
ax.set_title('Model I - GMRotD50 (g) T=1 s')
ax.set(ylabel='Off-fault distance (km)', xlabel='Along strike (km)')
#
cdt = (df['R_rup'] )
x, y, z = df['x'].values, df['y'].values, df['gm50_T1'].values
im = ax.tripcolor(x,y,z, cmap='RdGy_r', vmax=0.6)
##
con = ax.tricontour(x, y, df['R_rup'].values,[3,7,15], linewidths=0.5, colors='k')
plt.clabel(con, inline=1, fontsize=10, fmt='%.0f', colors='lime')
##
ax.plot([A[0], B[0]], [0.0, 0.0], c='k', alpha=0.7)
ax.scatter(x_hypo, 0.0, marker='*', c='gold')
fig.colorbar(im, shrink=0.75)
#
plt.tight_layout()
##

fig.savefig('/Users/elifo/Desktop/Model1_2dplot_Sa_1s.png', dpi=300)
```
