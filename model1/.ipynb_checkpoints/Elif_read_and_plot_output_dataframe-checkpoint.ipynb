## 2. Read previously prepared model outputs with dataframe df

```python
%matplotlib inline

import sys
sys.path.append('../')
import pandas as pd 
import numpy as np


df1 = pd.read_pickle('./DF_for_SCEC_benchmark')
df = df1.copy()

# final surface rupture extent to compute Rrup (station-to-fault distance)
# fault line segment: A to B
A = np.array([-5.0, 0.0])
B = np.array([30.0, 0.0])


# hypocenter
x_hypo = 16.0
```

## 3. Compute the closest distance to fault line (segment)

```python
from numpy import arccos, array, dot, pi, cross
from numpy.linalg import det, norm

# from: https://gist.github.com/nim65s/5e9902cd67f094ce65b0
def distance_numpy(A, B, P):
    """ segment line AB, point P, where each one is an array([x, y]) """
    if all(A == P) or all(B == P):
        return 0
    if arccos(dot((P - A) / norm(P - A), (B - A) / norm(B - A))) > pi / 2:
        return norm(P - A)
    if arccos(dot((P - B) / norm(P - B), (A - B) / norm(A - B))) > pi / 2:
        return norm(P - B)
    return norm(cross(A-B, A-P))/norm(B-A)
###


# R_rup = closest distance to fault plane
Rrup = []
xx = df['x']
yy = df['y']

for x,y in zip(xx, yy):
    P = np.array([x, y])
    Rrup.append(distance_numpy(A, B, P))
Rrup = np.array(Rrup)
df['R_rup'] = Rrup
#
```

### Use a custom distance range to compute standard deviation of simulation data 
This is to avoid scatter plot and show the varibility by bars at a given station spacing

```python
x = df['R_rup']
x_custom = []
interval = 0.1

_x_custom = np.arange(min(x), max(x)+interval, interval)
for _x in x:
    diff = (_x_custom-_x)**2
    cdt = (diff == min(diff))
#     print (_x, _x_custom[cdt])
    x_custom.append(_x_custom[cdt])
###

df['x_custom'] = np.array(x_custom)
df
```

## 4. Compare to GMPE (Mw =7)

```python
# READ GMPE
gmpe_file = '../m7_ss_ngaw2_bounds.csv'
gmpe = pd.read_csv(gmpe_file)  
# gmpe
```

```python
import matplotlib.pyplot as plt

def set_style(whitegrid=False, scale=0.85):
    import seaborn as sns
    
    sns.set(font_scale = scale)
    sns.set_style('white')
    if whitegrid: sns.set_style('whitegrid')
    sns.set_context('talk', font_scale=scale)
###


plt.close('all')
fig = plt.figure(figsize=(8, 4))
set_style(whitegrid=True)

ax = fig.add_subplot(121)
ax.set_title('T = 1 s')
ax.set_xlim(-1, 20.0)

x1, ymean1, std1, var1 = [], [], [], []
grouped_df = df.groupby('x_custom')
for key, item in grouped_df:
    values = item["gm50_T1"]
    x1.append(key)
    ymean1.append(np.mean(values))
    std1.append(np.std(values))  
    var1.append(np.var(values))
ax.errorbar(x1, ymean1, yerr=std1, fmt='o', ms=0, alpha=0.25,
             ecolor='k', elinewidth=1, capsize=1);
ax.scatter(x1, ymean1, c='salmon', ec='snow', alpha=0.9, lw=0.5, s=50)


ax.set_xlabel('Closest distance to fault (km)')
ax.set_ylabel('gm50 (g)')
# mean
R, Sa = gmpe['Rrup (km)'].values, gmpe['1.0 s Mean'].values
ax.plot(R, Sa, c='k', lw=0.5)
# min
R, Sa = gmpe['Rrup (km)'].values, gmpe['1.0 s Min'].values
ax.plot(R, Sa, c='k', alpha=1.0, linestyle='--', lw=0.5)
# max
R, Sa = gmpe['Rrup (km)'].values, gmpe['1.0 s Max'].values
ax.plot(R, Sa, c='k', alpha=1.0, linestyle='--', lw=0.5)
##




###
ax = fig.add_subplot(122)
ax.set_title('T = 3 s')
ax.set_xlim(-1, 20.0)

x3, ymean3, std3, var3 = [], [], [], []
grouped_df = df.groupby('x_custom')
for key, item in grouped_df:
    values = item["gm50_T3"]
    x3.append(key)
    ymean3.append(np.mean(values))
    std3.append(np.std(values))  
    var3.append(np.var(values))    
ax.errorbar(x3, ymean3, yerr=std3, fmt='o', ms=0, alpha=0.25,
             ecolor='k', elinewidth=1, capsize=1);
ax.scatter(x3, ymean3, c='blueviolet', ec='snow', alpha=0.75, lw=0.5, s=50)


ax.set_xlabel('Closest distance to fault (km)')
ax.set_ylabel('gm50 (g)')
# mean
R, Sa = gmpe['Rrup (km)'].values, gmpe['3.0 s Mean'].values
ax.plot(R, Sa, c='k', lw=0.5)
# min
R, Sa = gmpe['Rrup (km)'].values, gmpe['3.0 s Min'].values
ax.plot(R, Sa, c='k', alpha=1.0, linestyle='--', lw=0.5)
# max
R, Sa = gmpe['Rrup (km)'].values, gmpe['3.0 s Max'].values
ax.plot(R, Sa, c='k', alpha=1.0, linestyle='--', lw=0.5)
##


plt.tight_layout()
# fig.savefig('path_to_blabla.png', dpi=300)
plt.show()
```
