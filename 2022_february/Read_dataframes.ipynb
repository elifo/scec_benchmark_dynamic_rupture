# Read dataframe with all points

```python
%matplotlib inline

import pandas as pd 

# import pickle5 as pickle
# with open(picklename, "rb") as fh:
#     df1 = pickle.load(fh)


picklename = './outputs/DF_Model1_all_stations'
df1 = pd.read_pickle(picklename)

picklename = './outputs/DF_Model2_all_stations'
df2 = pd.read_pickle(picklename)

picklename = './outputs/DF_Model3_all_stations'
df3 = pd.read_pickle(picklename)
```

## Quick plots

### Model 1

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

df = df1

plt.subplot(131)
plt.scatter(df['R_rup'], df['gm50_T0p5'], c='darkturquoise', ec='snow', alpha=0.9, lw=0.5, s=50)
plt.xlabel('JB distance (km)')
plt.ylabel('GMROTD50')
plt.title('0.5 s')

plt.subplot(132)
plt.scatter(df['R_rup'], df['gm50_T1p0'], c='pink', ec='snow', alpha=0.9, lw=0.5, s=50)
plt.xlabel('JB distance (km)')
plt.ylabel('GMROTD50')
plt.title('1 s')

plt.subplot(133)
plt.scatter(df['R_rup'], df['gm50_T3p0'], c='peru', ec='snow', alpha=0.9, lw=0.5, s=50)
plt.xlabel('JB distance (km)')
plt.ylabel('GMROTD50')
plt.title('3 s')

plt.tight_layout()
```
